```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Unclean Datasets/WorldDependencyRatioData1960_2022.csv', skiprows=4)
del df['Indicator Name']
del df['Indicator Code']
del df['Unnamed: 67']


df.fillna(0, inplace=True)
df = df.convert_dtypes(convert_integer=True)

df['Country Name'] = df['Country Name'].str.upper()
df.rename(columns={'Country Name': 'Country'}, inplace=True)

yearList = [str(year)for year in np.arange(1960, 2023)]
df = pd.melt(df,id_vars='Country', value_vars=yearList, var_name="Year", value_name="dependency_ratio")
df
```

```python
df.to_csv('Cleaned Datasets/CleanedDepRatio1960_2022.csv', index=False)
```
